## Quick visual check on colour palettes

```python
import qstyles
from seaborn import palplot
import qstyles.palettes
```

```python
sequential_palettes = [
    "qs_violet_dark",
    "qs_violet_light",
    "qs_green_light",
    "qs_green_dark",
    "qs_plum_light",
    "qs_plum_dark",
    "psf_blue_dark",
    "psf_blue_light",
    "teal_light",
    "teal_dark",
]
base_palettes = [
    "grey_light",
    "grey_dark",
]
categorical_palettes = [
    "cat_colorful_dark",
    "cat_colorful_light",
    "cat_qs_dark",
    "cat_qs_light",
    "cat_bright",
]

for palette in sequential_palettes:
    palplot(getattr(qstyles.palettes.Sequential, palette))
for palette in base_palettes:
    palplot(getattr(qstyles.palettes.Base, palette))
```

```python
categorical_palettes = [
    "cat_colorful_dark",
    "cat_colorful_light",
    "cat_qs_dark",
    "cat_qs_light",
    "cat_bright",
]

for palette in categorical_palettes:
    palplot(list(getattr(qstyles.palettes.Categorical, palette).values()))
```

```python
diverging_palettes = ["qs_plum_green", "qs_green_violet"]
for palette in diverging_palettes:
    palplot(getattr(qstyles.palettes.Diverging, palette))
```
